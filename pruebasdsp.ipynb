```python
import google.generativeai as genai
import PIL.Image
import io

genai.configure(api_key="AIzaSyDXhX_KIrFNPkdy1JNeCbZK1EFFKSQ6iGc")

# Convertir imagen a bytes
img = PIL.Image.open('image.jpg').convert('RGB')
buffer = io.BytesIO()
img.save(buffer, format='JPEG')
image_bytes = buffer.getvalue()

# Crear el blob manualmente
image_blob = {
    "mime_type": "image/jpeg",
    "data": image_bytes
}

# Inicializa el modelo
model = genai.GenerativeModel('models/gemini-2.5-flash')

# Enviar la imagen como blob
response = model.generate_content([
    "Describe qué ves en esta imagen con detalle",
    image_blob
])

print(response.text)
```
